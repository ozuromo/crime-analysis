### Import

```python
import pandas as pd
import geopandas as gpd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='ticks')
sns.set_context('paper')
```

```python
df = gpd.read_file('../GENERATED-DATA/od_crime_census_by_tract.geojson')
df.shape, type(df)
```

### Feature Analysis / Visualization

```python
def geo_plt_col(df, col, scheme=None):
    f, ax = plt.subplots(figsize=(10, 10))

    df.plot(
        ax=ax,
        column=col,
        legend=True,
        scheme=scheme,
        cmap='viridis',
        edgecolor='black',
        linewidth=0.1,
        legend_kwds={
            'loc': 'lower right'
        },
        missing_kwds={
            "color": "lightgrey",
            "edgecolor": "red",
            "hatch": "///",
            "label": "Missing values",
        },
    )
    ax.set_axis_off()
    plt.show()
```

```python
ax = sns.histplot(
    data=df,
    x='idade',
    kde=True,
    binrange=(0, 100),
    binwidth=5,
    stat='percent',
)

ax.set_xlabel('Idade')
ax.set_ylabel('Percentual')

ax.set_xlim(0, 80)
ax.set_ylim(0, 20)
```

```python
geo_plt_col(df, 'CRIMES', scheme='quantiles')
```
